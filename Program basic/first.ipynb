```python
# Accept the marked price from the user and calculate the net amount as (Marked Price - Discount )to pay according to data
# >10000 ----------------20%
# >7000 and <=10000 -----15%
# <= 7000 ---------------10%


price =int(input("Please Enter the Marked price"))


if(price > 10000):
    price1=(price/100)*20
    price -=price1
    print("Net Amount to pay ",price)
    
if( price > 7000 and price <= 10000):
    price1=(price/100)*15
    price -=price1
    print("Net Amount to pay ",price)

if(price <= 7000 ):
    price1=(price/100)*10
    price -=price1
    print("Net Amount to pay ",price)

```
