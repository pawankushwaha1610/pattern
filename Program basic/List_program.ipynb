```python
#Write a Python program to check if each number is prime in a given list of numbers. Return True if all numbers are prime otherwise False.

lis=[0,5,7,13]

length=len(lis)

count=0

if 0 in lis or 1 in lis :
      print("no")
      

for item in lis:
      flag = True
      for i in range (2,item):
            if(item % 2==0):
                 flag = False
      if(flag):
            count+=1


if count == length:
      print("each number is prime")
else:
      print("no")
```
